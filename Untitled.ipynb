```python
import folium
```

```python
madrid = [40.4165000, -3.7025600]
m=folium.Map(
    location=madrid,
    tiles='Stamen Watercolor',
    zoom_start=12
)


for i in range (0,50,5):
    x=madrid[0]+(i/1000)
    y=madrid[1]+(i/1000)
    folium.Marker([x,y], popup='<i>Mt. Hood Meadows</i>', tooltip='hola').add_to(m)

m
    
```
